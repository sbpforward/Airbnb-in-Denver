```python
df_proptypeperhood = df_lst[['id','room_type','neighbourhood_cleansed','latitude','longitude']].copy()
df_proptypeperhood = df_proptypeperhood.groupby(['neighbourhood_cleansed','room_type']).size().to_frame('count').reset_index()
df_proptypeperhood.head(25)
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x="neighbourhood_cleansed", y="count", hue="room_type", data=df_proptypeperhood.copy())
```

```python
# Sum the total neighborhoods and find the top ten & Sort values, largest first

df_hoodtop10 = df_proptypeperhood.groupby(['neighbourhood_cleansed'])['count'].sum().sort_values(ascending=False)

df_hoodtop10
```

```python
# Top 10 Neighborhoods on airbnb

df_hoodtop10 = df_hoodtop10.iloc[0:10].reset_index()
```

```python
df_hoodtop10.head(10)
```

```python
#sns.color_palette("", 10)
ax = df_hoodtop10.plot.bar(x='neighbourhood_cleansed', y='count', rot=45, figsize=(12, 6), legend=False)
ax.set_xlabel("Neighborhood", fontsize=14)
ax.set_ylabel("# of Total Listings", fontsize=14)
plt.title('Most Available Neighborhoods on Airbnb in Denver', fontsize=20)
plt.tight_layout()
plt.savefig('most-available-airbnb-listings-in-denver.jpeg')
```

# Figuring out how to get the room types per neighborhood:

```python
top10neighborhoods = df_hoodtop10['neighbourhood_cleansed'].tolist()
```

```python
df_top10byroomtype = pd.DataFrame()

for neighborhood in top10neighborhoods:
    temp = df_proptypeperhood[df_proptypeperhood.neighbourhood_cleansed == neighborhood]
    #print(temp.head())
    df_top10byroomtype = df_top10byroomtype.append(temp)
        
```

```python
df_top10byroomtype.head(50)
```

```python
plt.figure(figsize=(16, 6))

ax = sns.barplot(x="neighbourhood_cleansed", y="count", hue="room_type", palette='YlOrRd', data=df_top10byroomtype.copy())
ax.set_xlabel("Neighborhood", fontsize=14)
ax.set_ylabel("Number of Listings", fontsize=14)
ax.set_title('Most Available Neighborhoods on Airbnb in Denver by Room Type', fontsize=20)
plt.gca().legend().set_title('Room Type')
plt.legend(loc=1, prop={'size': 14})

plt.tight_layout()
plt.savefig('most-available-airbnb-listings-in-denver-by-room-type.jpeg')
```

```python
# Interested in finding what the total numbers are for each in Denver

df_proptypeperhood = df_lst[['id','room_type','neighbourhood_cleansed','latitude','longitude']].copy()
df_roomtypetotals = df_proptypeperhood.groupby(['room_type']).size().to_frame('count').reset_index()
df_roomtypetotals.head()
```
